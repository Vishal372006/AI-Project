```python
import pandas as pd
```

```python
data = pd.read_csv("/content/Healthcare_Data(3).csv")
```

```python
data.head(30)
```

```python
x_data = data["Private beds distributed per 100,000 Population"].head(20)
y_data = data['Private beds available per 100,000 Population'].head(20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x_data,y_data)
plt.xlabel("Private beds distributed per 100,000 Population")
plt.ylabel("Private beds available per 100,000 Population")
plt.title("Private beds Distributed vs private beds avilable")
plt.show()
```
